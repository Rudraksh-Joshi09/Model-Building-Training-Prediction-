```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
# Load the iris dataset from CSV
df = pd.read_csv(r'D:\Notes\Internship\Assignment\iris.csv')

# Use only the numeric features (first 4 columns)
X = df.iloc[:, 0:4].values

#Step 1: Create and plot the dendrogram
linked = linkage(X, method='ward')  # 'ward' minimizes variance within clusters

plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=False,
           no_labels=True)
plt.title('Hierarchical Clustering Dendrogram (Iris Dataset)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Step 2: Perform Agglomerative Clustering (cutting the dendrogram into clusters)
hc = AgglomerativeClustering(n_clusters=3,linkage='ward')
labels = hc.fit_predict(X)

# Step 3: Plot the clustered data using Sepal Length vs Sepal Width
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Hierarchical Clustering (Agglomerative) on Iris Dataset')
plt.show()
```
