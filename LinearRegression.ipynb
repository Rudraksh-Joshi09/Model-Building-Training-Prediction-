```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

Dataframe = pd.read_csv(r"D:\Notes\Internship\Assignment\housing.csv")
Dataframe.head()
```

```python
x = Dataframe.iloc[:,:4].values  #independent Variable 
y = Dataframe.iloc[:,4].values   #Depended Variable
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=147)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
print(pred)
```

```python
print('actual Price \n',y_test,"\n")
print('Predected Price \n',pred)
```

```python
plt.scatter(y_test,pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```
